一、古希腊和罗马的经济思想  

1. 希腊时期的经济背景和经济思想  
    - 地中海沿岸，以城邦（city-state）为单位分布。奴隶制度  
    - 两个代表的民主城邦: 
        - 雅典：三面环海，地形狭长，山脉地形使其不能发展农业。以贸易为主（贸易型城邦）
        - 斯巴达：农业为生，全职武士，奴役“黑牢士”为其农耕。双王制  
    - 色诺芬《经济论》: 
        - 最早提出经济学————家庭管理
        - 重视农业，认为农业是财富的源泉
        - 分工
        - 客观价值和主观价值
        - 商业的重要性
        - 稳定和平政治环境的重要性
    - 古希腊三哲：苏格拉底、柏拉图、亚里士多德：  
        - 柏拉图：理想国：  
            1. 社会分工观（人生而不平等）：
                1. 农人、工匠（生产）
                2. 武士（保护）
                3. 哲学和理性的贵族（指引方向）
            2. 奴隶主之间的共产：
                1. 不应以今天的观点批判古人
                2. 武士的经济由国家供给，没有私有财产（共产观点）
            3. 鄙视商业与贸易
